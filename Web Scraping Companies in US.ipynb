```python
from bs4 import BeautifulSoup as bs
import requests
```

```python
url="https://en.wikipedia.org/wiki/List_of_largest_companies_in_the_United_States_by_revenue"
page=requests.get(url)
```

```python
soup=bs(page.text,"html")
```

```python
table=soup.find_all("table")[1]
```

```python
my_table=soup.find("table", class_="wikitable sorta")
```

```python
title=soup.find_all("th")[0:7]
title
```

```python
title_list=[i.text.strip() for i in title]
print(title_list)
```

```python
# for i in title_list:
#     print(i.strip("\n"))
import pandas as pd
```

```python
header=pd.DataFrame(columns=title_list)
```

```python
header
```

```python
tr=soup.find_all("tr")[1:]
tr
table_data_list=[]
for i in tr:
    row_data=i.find_all("td")
    table_data=[data.text.strip() for data in row_data]
    table_data_list.append(table_data)
```

```python
# del table_data_list[0]
# table_data_list=table_data_list.pop(0)
```

```python
pandas_table=pd.DataFrame(table_data_list)
pandas_table=pandas_table.drop(0)
```
